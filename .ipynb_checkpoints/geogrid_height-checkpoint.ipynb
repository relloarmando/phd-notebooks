```python
from netCDF4 import Dataset
import matplotlib.pyplot as plt
ds11, ds12, ds13 = Dataset('1m_geo_em.d01.nc'),  Dataset('1m_geo_em.d02.nc'),  Dataset('1m_geo_em.d03.nc')
ds21, ds22, ds23 = Dataset('30s_geo_em.d01.nc'), Dataset('30s_geo_em.d02.nc'), Dataset('30s_geo_em.d03.nc')
ds31, ds32, ds33 = Dataset('15s_geo_em.d01.nc'), Dataset('15s_geo_em.d02.nc'), Dataset('15s_geo_em.d03.nc')
print (ds1.file_format)
#print (dataset1.dimensions.keys())
#print (dataset1.variables.keys())
```

```python
hgt11, hgt21, hgt31 = ds11.variables['HGT_M'][0,:,:], ds21.variables['HGT_M'][0,:,:], ds31.variables['HGT_M'][0,:,:]
hgt12, hgt22, hgt32 = ds12.variables['HGT_M'][0,:,:], ds22.variables['HGT_M'][0,:,:], ds32.variables['HGT_M'][0,:,:]
hgt13, hgt23, hgt33 = ds13.variables['HGT_M'][0,:,:], ds23.variables['HGT_M'][0,:,:], ds33.variables['HGT_M'][0,:,:]
lat1, lon1 = ds1.variables['XLAT_M'][0,:,:], ds1.variables['XLONG_M'][0,:,:]
lat2, lon2 = ds2.variables['XLAT_M'][0,:,:], ds2.variables['XLONG_M'][0,:,:]
lat3, lon3 = ds3.variables['XLAT_M'][0,:,:], ds3.variables['XLONG_M'][0,:,:]
hgt1.shape
```

```python
import numpy as np 

f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(12.6,4))
ax1.contourf(hgt11, extent=[lon1.min(), lon1.max(), lat1.min(), lat1.max()])
ax2.contourf(hgt21, extent=[lon1.min(), lon1.max(), lat1.min(), lat1.max()])
ax3.contourf(hgt31, extent=[lon1.min(), lon1.max(), lat1.min(), lat1.max()])

plt.contourf(hgt11, extent=[lon1.min(), lon1.max(), lat1.min(), lat1.max()])
#f.subplots_adjust(right=1.5)
#plt.colorbar()


plt.show()
```

```python
import numpy as np 

f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(13,4))
ax1.contour(hgt13, extent=[lon3.min(), lon3.max(), lat3.min(), lat3.max()])
ax2.contour(hgt23, extent=[lon3.min(), lon3.max(), lat3.min(), lat3.max()])
ax3.contour(hgt33, extent=[lon3.min(), lon3.max(), lat3.min(), lat3.max()])

plt.show()
```
