```python
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from wrf import getvar, vinterp, ALL_TIMES, to_np, disable_xarray
dir='/mnt/sda2/exp2_201303_sodar/'
#dir='D:/exp2_201303_sodar'
ncg1 = Dataset(dir+'/gfs_pbl1_d3_20130301.nc')
ncg2 = Dataset(dir+'/gfs_pbl2_d3_20130301.nc')
ncn1 = Dataset(dir+'/nam_pbl1_d3_20130301.nc')
ncn2 = Dataset(dir+'/nam_pbl2_d3_20130301.nc')
```

```python
import numpy as np
import pandas as pd
wrftime = getvar(ncg1, "times", timeidx=ALL_TIMES)
utc2loc = np.timedelta64(6,'h')
wrftimes = pd.to_datetime(wrftime.values)-utc2loc

timeArraySod=np.arange(np.datetime64('2012-08-16 00:30'), np.datetime64('2013-04-11 00:30'),np.timedelta64(30,'m'))
print(wrftimes.size, wrftimes[0], wrftimes[wrftimes.size-1])
print(timeArraySod.size, timeArraySod[0], timeArraySod[timeArraySod.size-1])
```

```python
g1 = pd.read_pickle('exp2_gfs_pbl1_d3.pk')
g2 = pd.read_pickle('exp2_gfs_pbl2_d3.pk')
n1 = pd.read_pickle('exp2_nam_pbl1_d3.pk')
n2 = pd.read_pickle('exp2_nam_pbl2_d3.pk')
g1p = g1.interpolate(limit=99)
g2p = g2.interpolate(limit=99)
```

```python
import pandas as pd
# Variables to store data temporarily #238 dias * 48dt = 11424 times
# (11424 lineas de los tiempos, 98 niveles)
spd=(pd.read_csv('wind', header=None, low_memory=False)).values.reshape((11424, 98))
drc=(pd.read_csv('direc', header=None, low_memory=False)).values.reshape((11424, 98))
vte=(pd.read_csv('temp', header=None, low_memory=False)).values.reshape((11424, 98))

# (11424 lineas de los tiempos, 98 niveles)
spd[spd>99.9] = np.nan
drc[drc>999 ] = np.nan
colNames=["spd_"+str((i+3)*10) for i in range(98)] 
import numpy.ma as ma
#spd = np.where(np.isnan(spd), ma.array(spd, mask=np.isnan(spd)).mean(axis=0), spd) 
sp=pd.DataFrame(data=spd, index=timeArraySod, columns=colNames, dtype=None)
dd=pd.DataFrame(data=drc, index=timeArraySod, columns=colNames, dtype=None)
TV=pd.DataFrame(data=vte, index=timeArraySod, columns=colNames, dtype=None)
```

```python
fig = plt.figure(figsize=[18,6])
skip = 6

X = np.arange(331)[::skip]/skip
Y = 0
XX, YY = np.meshgrid(X, Y)

sw = g1['w'+level][::skip]
dw = np.radians(g1['wd'+level][::skip])
UX, VY = sw*np.sin(dw), sw*np.cos(dw)
plt.quiver(XX, YY, UX, VY, angles='xy', scale_units='xy', scale=1, width=0.0015,\
           color='mediumblue', label='w'+level+'_'+model)

ss = g1['s100'][::skip]
ds = np.radians(g1['sd'+level][::skip])
UX, VY = ss*np.sin(ds), ss*np.cos(ds)
plt.quiver(XX, YY, UX, VY, angles='xy', scale_units='xy', scale=1, width=0.0015,\
           color='k', label='s'+level)

sx, sy= X.size, 21
plt.yticks(np.linspace(-9,11,sy))
plt.xticks(X[::4], sw.index[::4], rotation=45);
plt.ylabel('Magnitud Vectores')

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.title('Vectores wrf_'+model+' y sodar a '+level+'m');

plt.savefig('error_plots/quiver_'+model+'_'+level+'m.png', dpi=300)
```

```python
fig, ax = plt.subplots(3, 1, sharex=True, figsize=[18,9])
sw = g1['w'+level]
dw = np.radians(g1['wd'+level])
UX, VY = sw*np.sin(dw), sw*np.cos(dw)
ax[0].plot(UX, label='U_'+model, color='mediumblue')
ax[1].plot(VY, label='V_'+model, color='mediumblue')

ss = g1['s'+level]
ds = np.radians(g1['sd'+level])
UX, VY = ss*np.sin(ds), ss*np.cos(ds)
ax[0].plot(UX, label='Usodar', color='black')
ax[1].plot(VY, label='Vsodar', color='black')

tv_sod = TV['spd_'+level]['2013-02-28 18':'2013-03-07 15:01']
ax[2].plot(tv_sod, label='tv_sod', color='k')
ax[2].plot(tvg1-273, label='tv_'+model+'_'+level, color='tomato')

ax[2].grid(True)
ax[2].legend()

ax[0].grid(True)
ax[0].legend()
ax[1].grid(True)
ax[1].legend();
plt.tight_layout()

ax[0].set_title('Componentes U  a '+level+'m')
ax[1].set_title('Componentes V  a '+level+'m')
ax[2].set_title('Temp Virtual a '+level+'m')
ax[0].set(ylabel='Wind U')
ax[1].set(ylabel='Wind V')
ax[2].set(ylabel='Degrees C')
ax[0].axhline(0,color='k', linewidth=1)
ax[1].axhline(0,color='k', linewidth=1)

plt.savefig('error_plots/componentes_'+model+'_'+level+'m.png', dpi=300)
```
