```python
lev = 8
# TODOS LOS EXPERIMENTOS WRF SEA BREEZE
sodmsk = ma.masked_invalid(sea) 
inv = sodmsk.mask
idx = ~inv
sodmsk.shape
test = 'wrf'
# TODOS LOS MODELOS UN SOLO NIVEL WRF1...WRF12
#print('TODOS LOS MODELOS NIVEL '+levlabel[lev])
#wrf = eval(exp[inds,lev][idx[:,lev]])
p = [eval(test+str(i)+'[inds,lev][idx[:,lev]]') for i in range(1,13)]
pm = [np.mean(p[i]) for i in range(12)] 
r = [sodmsk[:,lev][idx[:,lev]].data for i in range(12)]
rm = [np.mean(r[i]) for i in range(12)]
p_pm = [(p[i] - pm[i]) for i in range(12)]
r_rm = [(r[i] - rm[i]) for i in range(12)]
crmsd = [np.sqrt(np.average(np.square((p_pm[i]) - (r_rm[i])))) for i in range(12)]
sdevp = [np.std(p[i]) for i in range(12)]
sdevr = [np.std(r[i]) for i in range(12)]
sdev = [[sdevr[i], sdevp[i]] for i in range(12)];
ratio= [sdevp[i]/sdevr[i]  for i in range(12)]
ccorr =[ np.corrcoef(r[i], p[i])[1,0] for i in range(12)]
sdev = np.array(np.append(sdevr[0], sdevp))
crmsd = np.array(np.append(0, crmsd))
ccoef = np.array(np.append(1, ccorr))
print(test, levlabel[lev],  sdev[0], crmsd[0], ccoef[0], ratio[0])
```

```python
print(levlabel[lev], '\n RATIO', [("%.3f" % e) for e in ratio], 'CORR',  [("%.3f" % e) for e in ccorr],)
```
