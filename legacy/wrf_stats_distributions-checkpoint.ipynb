```python
def statswrf(model):
    e300, e200, e100, e50, e30 =model['s300']-model['w300'], \
         model['s200']-model['w200'], model['s100']-model['w100'], model['s50']-model['w50'], model['s30']-model['w30']
    b300, b200, b100, b50, b30 = e300.mean(), e200.mean(), e100.mean(),  e50.mean(),  e30.mean()
    mae300, mae200, mae100, mae50, mae30 = abs(e300).mean(), abs(e200).mean(),  abs(e100).mean(),  abs(e50).mean(),  abs(e30).mean()
    mse300, mse200, mse100, mse50, mse30 = (e300**2).mean(), (e200**2).mean(),  (e100**2).mean(),  (e50**2).mean(),  (e30**2).mean()
    rmse300, rmse200, rmse100, rmse50, rmse30 = mse300**.5, mse200**.5, mse100**.5, mse50**.5, mse30**.5
    cov300, cov200, cov100, cov50, cov30 = np.cov(model['s300'], model['w300']), np.cov(model['s200'], model['w200']), \
                                                       np.cov(model['s100'], model['w100']), np.cov(model['s50'], model['w50']), \
                                                       np.cov(model['s30'], model['w30'])
    corr300, corr200, corr100, corr50, corr30 = np.corrcoef(model['s300'], model['w300']), np.corrcoef(model['s200'], model['w200']), \
                                                       np.corrcoef(model['s100'], model['w100']), np.corrcoef(model['s50'], model['w50']), \
                                                       np.corrcoef(model['s30'], model['w30'])
    labels = ['AGL m', 'BIAS', 'MAE', 'RMSE', 'COV', 'R']
    stats = [('300m', b300, mae300, rmse300, cov300[0,1], corr300[0,1]),
         ('200m', b200, mae200, rmse200, cov200[0,1], corr200[0,1]),
         ('100m', b100, mae100, rmse100, cov100[0,1], corr100[0,1]),
         ('50m', b50, mae50, rmse50, cov50[0,1], corr50[0,1]),
         ('30m', b30, mae30, rmse30, cov30[0,1], corr30[0,1])]
   
    model = pd.DataFrame.from_records(stats, columns=labels)
    model.set_index('AGL m', inplace=True)

    pd.set_option('display.precision',3)
    return model
```

```python
# https://docs.scipy.org/doc/scipy-0.19.1/reference/generated/scipy.stats.exponweib.html
from scipy import stats
import matplotlib.pyplot as plt
def weibstat():
    
    # Fit Weibull function, some explanation below
    s300, s200, s100, s50, s30 = g1p['s300'], g1p['s200'], g1p['s100'], g1p['s50'], g1p['s30']
    sp300 = stats.exponweib.fit(s300, floc=0, f0=1)
    sp200 = stats.exponweib.fit(s200, floc=0, f0=1)
    sp100 = stats.exponweib.fit(s100, floc=0, f0=1)
    sp50 = stats.exponweib.fit(s50, floc=0, f0=1)
    sp30 = stats.exponweib.fit(s30, floc=0, f0=1)
    
    w300g1, w200g1, w100g1, w50g1, w30g1 = g1p['w300'], g1p['w200'], g1p['w100'], g1p['w50'], g1p['w30']
    w300g1 = stats.exponweib.fit(w300g1, floc=0, f0=1)
    w200g1 = stats.exponweib.fit(w200g1, floc=0, f0=1)
    w100g1 = stats.exponweib.fit(w100g1, floc=0, f0=1)
    w50g1 = stats.exponweib.fit(w50g1, floc=0, f0=1)
    w30g1 = stats.exponweib.fit(w30g1, floc=0, f0=1)
    
    w300g2, w200g2, w100g2, w50g2, w30g2 = g2p['w300'], g2p['w200'], g2p['w100'], g2p['w50'], g2p['w30']
    w300g2 = stats.exponweib.fit(w300g2, floc=0, f0=1)
    w200g2 = stats.exponweib.fit(w200g2, floc=0, f0=1)
    w100g2 = stats.exponweib.fit(w100g2, floc=0, f0=1)
    w50g2 = stats.exponweib.fit(w50g2, floc=0, f0=1)
    w30g2 = stats.exponweib.fit(w30g2, floc=0, f0=1)
    
    w300n1, w200n1, w100n1, w50n1, w30n1 = n1p['w300'], n1p['w200'], n1p['w100'], n1p['w50'], n1p['w30']
    w300n1 = stats.exponweib.fit(w300n1, floc=0, f0=1)
    w200n1 = stats.exponweib.fit(w200n1, floc=0, f0=1)
    w100n1 = stats.exponweib.fit(w100n1, floc=0, f0=1)
    w50n1 = stats.exponweib.fit(w50n1, floc=0, f0=1)
    w30n1 = stats.exponweib.fit(w30n1, floc=0, f0=1)
    
    w300n2, w200n2, w100n2, w50n2, w30n2 = n2p['w300'], n2p['w200'], n2p['w100'], n2p['w50'], n2p['w30']
    w300n2 = stats.exponweib.fit(w300n2, floc=0, f0=1)
    w200n2 = stats.exponweib.fit(w200n2, floc=0, f0=1)
    w100n2 = stats.exponweib.fit(w100n2, floc=0, f0=1)
    w50n2 = stats.exponweib.fit(w50n2, floc=0, f0=1)
    w30n2 = stats.exponweib.fit(w30n2, floc=0, f0=1)
    
    labels = ['AGL m', 's_shape', 's_scale', 'g1_shape', 'g1_scale', 'g2_shape', 'g2_scale', 'n1_shape', 'n1_scale', 'n2_shape', 'n2_scale']
    wstats = [('300m', sp300[1], sp300[3], w300g1[1], w300g1[3], w300g2[1], w300g2[3], w300n1[1], w300n1[3], w300n2[1], w300n2[3]),
              ('200m', sp200[1], sp200[3], w300g1[1], w300g1[3], w300g2[1], w300g2[3], w300n1[1], w300n1[3], w300n2[1], w300n2[3]),
              ('100m', sp100[1], sp100[3], w100g1[1], w100g1[3], w100g2[1], w100g2[3], w100n1[1], w100n1[3], w100n2[1], w100n2[3]),
              ('50m',  sp100[1], sp100[3], w50g1[1],  w50g1[3] , w50g2[1],  w50g2[3] , w50n1[1],  w50n1[3] , w50n2[1],  w50n2[3]) ,
              ('30m',  sp100[1], sp100[3], w30g1[1],  w30g1[3] , w30g2[1],  w30g2[3] , w30n1[1],  w30n1[3] , w30n2[1],  w30n2[3])]
    
    df = pd.DataFrame.from_records(wstats, columns=labels)
    df.set_index('AGL m', inplace=True)
    pd.set_option('display.precision',3)
    return df
```
