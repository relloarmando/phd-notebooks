```python
def hovtSBu(xp,tt,angle,ztop):
    plt.figure(figsize=(14,6))  
    
    fd ='../../../../../d03/'+expruns[xp]+'.nc'
    ncfile = Dataset(fd)
  
    uv, wd = getvar(ncfile, "wspd_wdir", units="m s-1", timeidx=tt)   
    ua = getvar(ncfile, "ua", units="m s-1", timeidx=tt)
    wa = getvar(ncfile, "wa", units="m s-1", timeidx=tt)
    tv = getvar(ncfile, "tv", units="degC", timeidx=tt)
    z = getvar(ncfile, "z", units="m", timeidx=tt)

    # Define a pivot point and angle in grid coordinates
    pivot_point = CoordPair(x=22, y=22)

    # Calculate the vertical cross section.
    uv_vert = vertcross(uv, z, pivot_point=pivot_point, angle=angle, latlon=True)[:ztop,:]
    wd_vert = vertcross(wd, z, pivot_point=pivot_point, angle=angle, latlon=True)[:ztop,:]
    u_vert = uv_vert*np.cos(np.radians(wd_vert-angle))
    w_vert = vertcross(wa, z, pivot_point=pivot_point, angle=angle, latlon=True)[:ztop,:]
    tv_vert = vertcross(tv, z, pivot_point=pivot_point, angle=angle, latlon=True)[:ztop,:]
    z_vert = tv_vert.vertical.values
    coord_pairs = to_np(u_vert.coords["xy_loc"])
    size = coord_pairs.size

    i=0
    lats, lons = {}, {}
    for i in range(size):
        lats[i] = tv_vert.xy_loc.values[i].lat
        lons[i] = tv_vert.xy_loc.values[i].lon
        
    skip = 1
    lons = np.array(list(lons.values()))[::skip]
    LONS, ZHOV = np.meshgrid(lons, z_vert)
    
    plt.contourf(LONS, ZHOV, np.abs(u_vert), cmap='rainbow')  #, levels=levels, vmin=vmin, vmax=vmax,
    plt.colorbar();
    
    vmin, vmax = 20, 35
    levels = np.arange(vmin, vmax, 1)
    CS = plt.contour(LONS, ZHOV, tv_vert, colors='w', levels=levels)    
    plt.clabel(CS, inline=1, fmt='%2.1f', colors='w', fontsize=10)
    plt.gca().patch.set_color('springgreen')

    Q = plt.quiver(LONS[:,::2], ZHOV[:,::2], u_vert[:,::2], w_vert[:,::2], pivot='mid', scale=100, width=0.0030)
    plt.quiverkey(Q, X=0.85, Y=1.05, U=5, label=' 5 m/s', labelpos='E')
    
    timestr = ua.Time.values - np.timedelta64(6,'h')
    ts = pd.to_datetime(str(timestr)) 
    d = ts.strftime('%Y%m%d_%H_%M')
    plt.title(expnames[xp]+' U Speed and Temp for Sea Breeze\n rotated '+str(angle)+' deg to Sea Breeze axis '+d);
    
    # Set the x-ticks to use latitude and longitude labels.
    xticks, xlabel = plt.xticks()
    x_labels = [pair.latlon_str(fmt="{:.2f}, {:.2f}") for pair in to_np(coord_pairs)]
    xl = np.array(x_labels)
    sx, sxl = xticks.size,  int(xl.size)
    x1, x2, dx = xticks[1], xticks[sx-1], int(sxl/(sx-1))
    plt.xticks(xticks[1:sx-1], xl[dx:sxl:dx], rotation=0)

    #plt.savefig('figs/Sea Breeze T Axis'+d+'.png', dpi=300)
```
