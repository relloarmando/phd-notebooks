```python
def vstats(p,r):
    rm = np.ma.masked_invalid(r)
    rm[rm.mask] = 0
    p[rm.mask] = 0
    rmean = np.mean(rm, axis=0)
    pmean = np.mean(p, axis=0)
    rmax = np.max(rm, axis=0)
    pmax = np.max(p, axis=0)
    rstd = np.nanstd(rm, axis=0)
    pstd = np.nanstd(p, axis=0)
    '''
    Calculates the centered root-mean-square (RMS) difference between 
    two variables PREDICTED and REFERENCE (E'). The latter is calculated 
    using the formula:
    (E')^2 = sum_(n=1)^N [(p_n - mean(p))(r_n - mean(r))]^2/N
    where p is the predicted values, r is the reference values, and
    N is the total number of values in p & r. Note that p & r must
    have the same number of values.   
    '''
    crmsd = np.square((p - pmean) - (rm - rmean))
    crmsd = np.sum(crmsd, axis=0)/(p.size/12)
    crmsd = np.sqrt(crmsd)
    crmsd = crmsd.data
    corrp = np.ma.corrcoef(p,rm)[1,0]
    
    return pmean, rmean, pmax, rmax, rstd, pstd, crmsd
```
