```python
import math  
import numpy as np
import matplotlib.pyplot as plt
k=3             # 2 ciclos por tiempo de 128
T=128
t=np.arange(1,129)
pi=math.pi
u=2*pi*(k/T)*t          # funcion u

fig = plt.figure()
fig.subplots_adjust(right=1.5)

ax1 = fig.add_subplot(221)
plt.plot(np.cos(u))
plt.ylabel(r'$\cos u$')
plt.xlabel(r'$t$')

ax2 = fig.add_subplot(222)
plt.plot(np.sin(u), color='r')
plt.ylabel(r'$\sin u$')
plt.xlabel(r'$t$')

print('En general las graficas pueden tener k ciclos y amplitud A')
print('A*cos(2*pi(k/T)*t)')
```

```python
import pandas as pd
import numpy.ma as ma

# Variables to store data temporarily #238 dias * 48dt = 11424 times
# (11424 lineas de los tiempos, 98 niveles)
spd=(pd.read_csv('python/wind', header=None, low_memory=False)).values.reshape((11424, 98))
drc=(pd.read_csv('python/direc', header=None, low_memory=False)).values.reshape((11424, 98))

# (11424 lineas de los tiempos, 98 niveles)
spd[spd>99.9] = np.nan
drc[drc>999 ] = np.nan
spd = np.where(np.isnan(spd), ma.array(spd, mask=np.isnan(spd)).mean(axis=0), spd)
colNames=["spd_"+str((i+3)*10) for i in range(98)] 
timeArray=np.arange(np.datetime64('2012-08-16 00:30'), np.datetime64('2013-04-11 00:30'),np.timedelta64(30,'m'))
sp=pd.DataFrame(data=spd, index=timeArray, columns=colNames, dtype=None)
dd=pd.DataFrame(data=drc, index=timeArray, columns=colNames, dtype=None)
```
